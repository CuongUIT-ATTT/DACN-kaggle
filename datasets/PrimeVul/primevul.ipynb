```python
import os
import json
import pandas as pd

# Go two directories up
os.chdir("..")
os.chdir("..")
```

```python

# File paths
file_paths = [
    "datasets/cwe20cfa/cwe20cfa_train.jsonl",
    "datasets/cwe20cfa/cwe20cfa_valid.jsonl",
    "datasets/cwe20cfa/cwe20cfa_test.jsonl"
]

# Accumulator for all records
all_records = []

for path in file_paths:
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line:  # skip empty lines
                try:
                    all_records.append(json.loads(line))
                except json.JSONDecodeError as e:
                    print(f"Skipping malformed line in {path}: {e}")

# Convert to DataFrame
original_cwe20cfa_complete = pd.DataFrame(all_records)

# Preview
original_cwe20cfa_complete.head()
```

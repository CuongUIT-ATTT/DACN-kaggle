# Pearson R

```python
import pandas as pd

scmetrics = pd.read_csv("benchmarks\scmetrics_test.csv", index_col=0)
scmetrics
```

Compute the Pearson r between WGA_2 and Neighbor Purity

```python
import numpy as np

wga2   = scmetrics["wga_k2"]
purity = scmetrics["n_purity"]

scmetrics[["wga_k2", "n_purity"]]
```

```python
# Remove outlier 100/0
pearson = np.corrcoef(wga2[1:], purity[1:])[0,1]
print(f"r: {pearson:.2f}")
```
